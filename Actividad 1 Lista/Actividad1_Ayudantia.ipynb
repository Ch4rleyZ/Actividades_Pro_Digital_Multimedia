## `Estructura if-else`

````python
numero=2 #Declaramos una variable con algún valor

if numero == 5: # Si la varaible es igual al valor 5:
    print("numero es igual a 5") #Muestra por pantalla que es igual a 5
else: #Si No
    print("numero es igual a", numero) # Muestra por pantalla que el numero es igual al numero dado
````

````python
numero1=2
numero2=2
if numero1 and numero2 == 2:
    print("Es dos")
else: 
    print("es otro numero")
````

## `Ejemplo con operadores logicos`

````python
# Operador AND
expresion1 = 2 < 3
expresion2 = 4 > 2


if expresion1 and expresion2: #   Si la condicion 1 y la condicion 2 son verdaderas :
  print("Se cumplen ambas expresiones")
else: # Si una de las dos es falsa:
  print("Una de las expresiones no se cumple")
````

## `Sentencia if-elif-else`

````python
num = int(input("Ingrese la edad: "))

type(num)
if num < 18:
    print("No puede ver la película")
elif 18 <= num <= 60:
    print("Puede ver la película con descuento")
else:
    print("Puede ver la película")
````

## `Que es Matplotlib`

**Matplotlib es una biblioteca de Python para crear gráficos y visualizaciones de datos. Se complementan correctamente numpy y matplot**

["Lo Basico de Matplotlib"](https://www.w3schools.com/python/matplotlib_pyplot.asp)

`Pero, ¿cómo graficar la función anterior?`

````python
import matplotlib.pyplot as plt #Apodo por convencion plt
import numpy as np

x=np.linspace(0,10,100) # 100 puntos linealmente espaciados de 0 a 10
y=np.sin(x) #Funcion Seno

plt.plot(x,y, color="red", label="Funcion seno")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.title("Funcion seno")
plt.legend()
plt.grid() #por defecto es 1 = True
plt.show()
````

## `Deben proporcionar una cantidad suficiente de puntos para que la función quede bien definida. Esto es similar a la tasa de muestreo: a mayor cantidad de muestras, la representación discreta de la función será más precisa y fiel a la original.`

````python
import matplotlib.pyplot as plt #Apodo por convencion plt
import numpy as np

x=np.linspace(0,10,1000) # 1000 puntos linealmente espaciados de 0 a 10
y=np.sin(x) #Funcion seno

plt.plot(x,y, color="red", label="Funcion seno")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.title("Funcion seno")
plt.legend()
plt.grid() #por defecto es 1 = True
plt.show()
````

````python
#Funcion exponencial


x=np.linspace(-10,10,1000) # 100 puntos linealmente espaciados de 0 a 10
y=np.exp(x)


plt.plot(x,y, color="red", label="Funcion exponencial")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.title("Funcion exponencial")
plt.legend()
plt.grid() #por defecto es 1 = True
plt.show()
````

## `Mas avanzado`

`Dandole uno mismo los parametros mas personalizados`

$$A*sen(2*pi*f*t)$$

````python
import numpy as np
import matplotlib.pyplot as plt

amplitud= 10
frecuencia= 1 #Hz
tiempo=np.linspace(0,5,1000)

funcion= amplitud*np.cos(2*np.pi*frecuencia*tiempo)

plt.figure(figsize=(10,5)) # Modificar el tamaño del grafico
plt.plot(tiempo,funcion)
plt.title("Funcion coseno")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.grid(1)
plt.show()
````

## `Si queremos crear subplots`



````python
import numpy as np
import matplotlib.pyplot as plt

t=np.linspace(0,10,100)
funcion1=np.sin(t)
funcion2=np.cos(t)


plt.figure(figsize=(10,5)) # Modificar el tamaño del grafico
plt.subplot(1,2,1)  # Fila, Columna, numero grafico
plt.plot(funcion1)
plt.title("Funcion seno")
plt.grid()

plt.subplot(1,2,2) # Segundo grafico
plt.plot(funcion2)
plt.title("Funcion coseno")
plt.grid()
````

## `Si queremos superponer funciones en un solo grafico`

````python
import numpy as np
import matplotlib.pyplot as plt

t=np.linspace(0,10,100)
funcion1=np.sin(t)
funcion2=np.cos(t)

plt.figure(figsize=(10,5)) # Modificar el tamaño del grafico
plt.plot(funcion1)
plt.plot(funcion2)
plt.title("Funciones")
plt.grid()
````
