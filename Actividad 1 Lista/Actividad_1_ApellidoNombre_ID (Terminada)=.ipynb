### 3.1 Ruido Gaussiano
Debes agregar la señal de ruido a la que creaste previamente. Adicionalmente, debes mostrar una gráfica con el resultado.

$$\frac{1}{\sqrt{2\pi} \cdot \sigma} \cdot e^{-\frac{(x - \mu)^2}{2\sigma^2}}$$

```python
def gaussian(x, mu, sig):
    return (
        1.0 / (np.sqrt(2.0 * np.pi) * sig) * np.exp(-np.power((x - mu) / sig, 2.0) / 2)
    )


x_values = np.linspace(-4, 4, 150)

mu=0
sigma=1
formula=gaussian(x_values,mu,sigma)

plt.plot(x_values,formula,label=f"Mu:{mu} Sigma={sigma}",color="red")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.title("Distribucion Gaussiana")
plt.legend()
plt.grid(True)
```

### 3.2 Ruido Rosa
Investiga en que consiste el ruido rosa, cual es la diferencia del ruido gaussiano y demuéstralo.

# `Explicación del Ruido Rosa y Ruido Gaussiano`

Esta explicación detalla qué es el **ruido rosa**, cómo se diferencia del **ruido gaussiano** (especialmente cuando se habla de ruido blanco gaussiano) y destaca sus principales características.



## Ruido Rosa

- **Definición:**  
  El ruido rosa es una señal cuyo contenido energético se distribuye de manera que la densidad espectral de potencia disminuye inversamente proporcional a la frecuencia (se sigue una ley 1/f).

- **Características:**
  - **Energía por octava constante:**  
    Cada rango de frecuencia, al duplicar la frecuencia (una "octava"), aporta aproximadamente la misma cantidad de energía.
  - **Aplicaciones naturales y en audio:**  
    Se observa en muchos procesos naturales y se utiliza en aplicaciones de audio, ya que su distribución de energía resulta más equilibrada para el oído humano.



## Ruido Gaussiano (Blanco)

- **Definición:**  
  El ruido gaussiano se caracteriza por tener sus valores distribuidos de acuerdo a una distribución normal (la clásica campana de Gauss). Cuando se combina con un espectro de potencia plano, se le denomina ruido blanco gaussiano.

- **Características:**
  - **Distribución normal:**  
    La amplitud de la señal sigue la distribución gaussiana.
  - **Densidad espectral plana:**  
    La potencia se distribuye uniformemente a lo largo de todas las frecuencias, sin favorecer ninguna en particular.



```python
import numpy as np
import matplotlib.pyplot as plt

def generate_white_noise(n):
    """Genera ruido blanco gaussiano de longitud n."""
    return np.random.normal(0, 1, n)

def generate_pink_noise(n):
    """
    Genera una aproximación de ruido rosa.
    Se parte de ruido blanco, se transforma al dominio de la frecuencia,
    se aplica un escalado inverso proporcional a sqrt(f) y se vuelve al dominio del tiempo.
    """
    # Generar ruido blanco
    white = np.random.normal(0, 1, n)
    # Transformada de Fourier
    f = np.fft.rfft(white)
    # Obtener el array de frecuencias
    freqs = np.fft.rfftfreq(n, d=1.0)
    # Crear un factor de escalado: evitar división por cero en la frecuencia 0
    scaling = np.ones_like(freqs)
    scaling[1:] = 1 / np.sqrt(freqs[1:])
    # Aplicar el escalado en el dominio de la frecuencia
    f = f * scaling
    # Volver al dominio del tiempo con la transformada inversa
    pink = np.fft.irfft(f, n)
    # Normalizar para tener media cero y varianza 1
    pink = (pink - np.mean(pink)) / np.std(pink)
    return pink

def compute_psd(signal):
    """
    Calcula la densidad espectral de potencia (PSD) de una señal.
    """
    psd = np.abs(np.fft.rfft(signal))**2
    freqs = np.fft.rfftfreq(len(signal), d=1.0)
    return freqs, psd

# Parámetros
n = 2**12  # Número de muestras

# Generar señales
white_noise = generate_white_noise(n)
pink_noise = generate_pink_noise(n)

# Calcular la PSD de cada señal
freqs_white, psd_white = compute_psd(white_noise)
freqs_pink, psd_pink = compute_psd(pink_noise)

# Graficar las señales en el dominio del tiempo y su PSD en dominio de frecuencia
plt.figure(figsize=(12,8))

# Ruido blanco (gaussiano) en el tiempo
plt.subplot(2,2,1)
plt.plot(white_noise)
plt.title("Ruido Blanco (Gaussiano) - Tiempo")
plt.xlabel("Muestras")
plt.ylabel("Amplitud")

# Ruido rosa en el tiempo
plt.subplot(2,2,2)
plt.plot(pink_noise)
plt.title("Ruido Rosa - Tiempo")
plt.xlabel("Muestras")
plt.ylabel("Amplitud")

# PSD del ruido blanco (log-log)
plt.subplot(2,2,3)
plt.loglog(freqs_white[1:], psd_white[1:])
plt.title("PSD Ruido Blanco (Gaussiano)")
plt.xlabel("Frecuencia")
plt.ylabel("Potencia")

# PSD del ruido rosa (log-log)
plt.subplot(2,2,4)
plt.loglog(freqs_pink[1:], psd_pink[1:])
plt.title("PSD Ruido Rosa")
plt.xlabel("Frecuencia")
plt.ylabel("Potencia")

plt.tight_layout()
plt.show()
```
